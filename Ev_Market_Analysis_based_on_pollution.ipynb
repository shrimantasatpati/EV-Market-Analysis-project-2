```python
state_temp = df[['Temp', 'state']]
```

```python
# Create a barplot
plt.barh(state_temp['state'], state_temp['Temp'], color='lightgray')

# Set the title, x-axis label, and y-axis label
plt.title('Temperature by State')
plt.xlabel('Temperature (F)')
plt.ylabel('States')
```

```python
plt.figure(figsize=(8,16));
sns.jointplot(x='status', y='state', data=df, kind='scatter',space=0.2,palette="coolwarm");
```

```python
plt.figure(figsize=(34,8));
sns.swarmplot(x="state", y="status", data=df)
```

```python
# Create a boolean mask for the poor status data
poor_status_mask = df['status'] == 'POOR'

# Select the state column from the dataset
state_column = df['state']

# Filter the state column using the boolean mask
poor_states = state_column[poor_status_mask]

# Print the state names
poor_states
```
